````python
def calc_fuel_prop(df):
    
    fuels = df['fuel'].unique()
    for fuel in fuels:
        total = df.shape[0]
        print('Combustível: {0} Quantidade: {1}'.format(
                fuel, df.query("fuel == @fuel")['fuel'].count()/total))

print("08")
calc_fuel_prop(df_08)
print("18")
calc_fuel_prop(df_18)
````

### Q2: Qual foi o nível de melhoria das classes de veículo no quesito "economia de combustível"?

````python
def calc_veh_fuel(df1, df2):
    
    classes = df2['veh_class'].unique()
    for v_class in classes:
        print('Classe: {0} Variação: {1}'.format(
                v_class, df1.query("veh_class == @v_class")['veh_class'].count()/
                         df2.query("veh_class == @v_class")['veh_class'].count()))

calc_veh_fuel(df_08, df_18)
````

````python
# carregue conjuntos de dados

df_08 = pd.read_csv('clean_08.csv')
df_18 = pd.read_csv('clean_18.csv')
````

### 9.1 Renomeie as colunas de 2008 para distingui-las das de 2018 após a combinação

````python
# renomeie colunas de 2008

col_list = df_08.columns
df_08.rename(columns=lambda x: x[:10]+"_2008", inplace=True)
````

````python
# exiba para verificar nomes
df_08.head()
````

### 9.2 Executar combinação interna
Para responder à última pergunta, só temos interesse em como o mesmo modelo de carro foi atualizado e como o mpg do novo modelo pode ser comparado ao do modelo antigo.

Execute uma combinação interna com a esquerda em “model_2008” e a direita em “model”


````python
df_18.head()
````

````python
# mescle os conjuntos de dados
df_combined = df_08.merge(df_18,how='left', left_on='model_2008', right_on='model', indicator=True)
````

````python
# exiba para verificar a mesclagem
df_combined.head()
display(df_combined)
````
