```python
# importa e carrega dados
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
#plt.style.use

df = pd.read_csv('../doc/powerplant_data.csv')
```

```python
# verificar se possui valores ausentes
def valores_ausentes(data):
    result = data.isnull().sum()
    display(result)
    
valores_ausentes(df)
```

```python
# confirmando com info()
df.info()
```

```python
# verifique se há dados duplicado
sum(df.duplicated())
```

```python
# linhas duplicadas
df[df.duplicated()]
```

```python
#emininando dados duplicados
df.drop_duplicates(inplace=True)
```
